#### master.csv

```python
### pitcher ID list 
pit_ID = []
pit_retro_ID = []
for pitcher_ID in total_KBO_final["pitcher_ID"]:
    if pitcher_ID == '':
        pass
    if pitcher_ID not in pit_ID:
        pit_ID.append(pitcher_ID)
        pit_retro_ID.append(str(pitcher_ID) + 'abc')
```

```python
### batter ID list 
bat_ID = []
bat_retro_ID = []
for batter_ID in total_KBO_final["batter_ID"]:
    if batter_ID == '':
        pass
    if batter_ID not in bat_ID:
        bat_ID.append(batter_ID)
        bat_retro_ID.append(str(batter_ID) + 'abc')
```

```python
playerID = pit_ID + bat_ID
retroID = pit_retro_ID + bat_retro_ID
```

```python
master = pd.DataFrame()
master['playerID'] = playerID
master['retroID'] = retroID
```
